```python
from langchain_core.messages import HumanMessage, AIMessage
```

```python
message_h = HumanMessage(content='''I've recently adopted a dog and I'm looking for some unique and creative name suggestions. Can you help me come up with a list of potential names that reflect different personalities and traits?''')
message_ai_dog = AIMessage(content='''Oh, absolutely! Because naming a dog is obviously the most important decision you'll ever make. Here’s a list of names that might just blow your mind:

1. **Sir Barksalot** - For the pup who thinks he’s royalty.''')

message_h_cat = HumanMessage(content = '''I've recently adopted a cat. Can you suggest some cat names ?''')
message_ai_cat = AIMessage(content = '''Absolutely! Here’s a list of unique and creative cat names that reflect different personalities and traits:

### Playful and Energetic
1. **Ziggy**
2. **Biscuit**
3. **Nibbles**''')

message_h_fish = HumanMessage(content = '''I've recently adopted a fish. Can you suggest some fish names ?''')
```

```python
chat = ChatOpenAI(model_name = 'gpt-4o-mini', 
                  model_kwargs ={'seed':365},
                temperature=0,
                max_tokens=50)
```

```python
response = chat.invoke([message_h,message_ai_cat, message_h_cat, message_ai_cat, message_h_fish]) 
```

```python
print(response.content)
```
